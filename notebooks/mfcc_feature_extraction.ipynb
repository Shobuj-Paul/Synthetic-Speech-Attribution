```python
import pandas as pd
import numpy as np
import librosa
import librosa.display
import librosa
import matplotlib.pyplot as plt
import IPython.display
```

```python
# Loading the data paths
df_test, df_dev, df_train = pd.read_csv('../assets/csv/test.csv'), pd.read_csv('../assets/csv/dev.csv'), pd.read_csv('../assets/csv/train.csv')
files_test, files_dev, files_train = df_test['track'], df_dev['track'], df_train['track']
labels_test, labels_dev, labels_train = df_test['algorithm'], df_dev['algorithm'], df_train['algorithm']
```

```python
file = 120
y, sr = librosa.load(files_test[file], sr=librosa.get_samplerate(files_train[file]))
label = labels_test[file]
t = np.linspace(0, len(y)/sr, len(y))
print("Total Time of Audio: {}s".format(round(t[-1], 2)))
print("Sampling Rate: {}Hz".format(sr))
librosa.display.waveplot(y, sr=sr)
IPython.display.Audio(files_test[file])
```

```python
mfcc = librosa.feature.mfcc(y=y, sr=sr)
print("Shape of MFCC: {}".format(mfcc.shape))
image = librosa.display.specshow(mfcc, x_axis='time', y_axis='mel')
```
